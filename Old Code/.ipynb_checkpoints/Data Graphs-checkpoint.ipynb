```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot(df):
    std = df.std(axis=1, numeric_only=True)
    mean = df.mean(axis=1, numeric_only=True)
    plt.scatter(mean, std, marker='.')
    plt.xlabel("Mean")
    plt.ylabel("Standard Deviation")
    plt.show()
    print(std.median())
    plt.hist(std, 50)
    plt.xlabel("standard deviation")
    plt.ylabel("number of genes")
    plt.show()
    print(mean.median())
    plt.hist(mean, 50)
    plt.xlabel("mean")
    plt.ylabel("number of genes")
    plt.show()
```
