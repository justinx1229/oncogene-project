```python
# normalize dataframe to 0
def normalize(dataframe):
    df = dataframe.copy()
    mean = df.iloc[:, 1:].mean(axis=1, numeric_only=True)
    df.iloc[:, 1:] = df.iloc[:, 1:].sub(mean, axis=0)
    return df
```
